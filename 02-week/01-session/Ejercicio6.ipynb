```python
class Notificacion:
    def preparar(self): pass
    def enviar(self): pass

class Email(Notificacion):
    def __init__(self, asunto, destinatario, cuerpo):
        self.asunto, self.destinatario, self.cuerpo = asunto, destinatario, cuerpo
    def preparar(self): return f"Email â†’ {self.destinatario} | Asunto: {self.asunto}"
    def enviar(self): return f"Enviado a {self.destinatario}"

class SMS(Notificacion):
    def __init__(self, numero, mensaje):
        self.numero, self.mensaje = numero, mensaje
    def preparar(self): return "SMS muy largo" if len(self.mensaje) > 160 else f"SMS â†’ {self.numero}"
    def enviar(self): return f"Enviado a {self.numero}"

class Push(Notificacion):
    def __init__(self, dispositivo, contenido):
        self.dispositivo, self.contenido = dispositivo, contenido
    def preparar(self): return f"Push â†’ {self.dispositivo}"
    def enviar(self): return f"Push enviado"

notificaciones = []
print("Sistema de Notificaciones")
while True:
    t = input("\nTipo (email/sms/push/salir): ").lower()
    if t == "email":
        notificaciones.append(Email(input("Asunto: "), input("Destinatario: "), input("Cuerpo: ")))
    elif t == "sms":
        notificaciones.append(SMS(input("NÃºmero: "), input("Mensaje: ")))
    elif t == "push":
        notificaciones.append(Push(input("Dispositivo: "), input("Contenido: ")))
    elif t == "salir":
        break

print("\n Procesando...")
for n in notificaciones:
    print(n.preparar(), "â†’", n.enviar())
```
